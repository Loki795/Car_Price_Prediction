```python
#we have feature called as year, so if one of the year is less like 1990 so the price will go far less, so we will derive a new
#feature using this year feature. Ex: 2020(currentyearvalue(what yr u need)) - 2014(yearvalue )
```

```python
df.columns
```

```python
finaldataset=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
#we didnt pick car name because that feature doesnt hold much value when compared with other for our business problem
```

```python
finaldataset.head()
```

```python
finaldataset['currentyear']= 2020
```

```python
finaldataset.head()
```

```python
finaldataset['Noofyear']= finaldataset['currentyear']- finaldataset['Year']
```

```python
finaldataset.head()
```

```python
finaldataset.drop(['Year','currentyear'], axis=1, inplace= True)
```

```python
finaldataset.head()
```

```python
#'Year''currentyear' both feature are dropped as we have created a new feature noofyear we dont need them
```

```python
#now we use dummy variable to create dummy for discrete features
```

```python
finaldataset= pd.get_dummies(finaldataset, drop_first=True)
```

```python
finaldataset.head()
```

```python
#dummy varibale is created we can see fuel_type CNG is missing because we have mentioned to drop first column
#whenever the fuel dummy both represent 0 0 it will be considere as cng thats why we remove the first column
```
